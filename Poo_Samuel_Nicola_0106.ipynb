# lenguaje python, crear la clase "animal" con los atributos "nombre", "raza" crear el contructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para simular el ladrido
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!🐱‍👤")

    # Función (método) para simular la carrera
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.😊")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 😜")
# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto pitbull con datos proporcionados
pitbull = Animal("Rocky", "Pitbull")

# Creamos el objeto chihuahua con datos proporcionados
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeton perro callejero con datos
perrocallejero = Animal("solovino", "Chafa")

# --- Uso de los objetos ---

print(f"Datos del primer animal: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

print(f"Datos del segundo animal: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

print(f"Datos del tercer animal: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print(" programa creado por Samuel Nicola NL = 40")
```
